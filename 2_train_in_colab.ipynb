```python
from sklearn.model_selection import train_test_split

img_list = glob('/content/mount/MyDrive/Side_Project/Driving_Control_System/Data/train/*.jpg') # ÌïôÏäµÏãúÌÇ¨ Ïù¥ÎØ∏ÏßÄ 5Ï≤úÏû• Î¶¨Ïä§Ìä∏
# train test split
train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=129) # 8:2

print(f"trainÌè¥Îçî ÏïàÏóê ÌååÏùº Í∞ØÏàò : {len(glob('/content/mount/MyDrive/Side_Project/Driving_Control_System/Data/train/*'))} Ïû•")
print(f'img_list : {len(img_list)} Ïû•')
```

```python
with open('/content/mount/MyDrive/Side_Project/Driving_Control_System/Data/train.txt', 'w') as f: # train 4Ï≤úÏû• Î¶¨Ïä§Ìä∏
    f.write('\n'.join(train_img_list) + '\n')
with open('/content/mount/MyDrive/Side_Project/Driving_Control_System/Data/val.txt', 'w') as f: # val 1Ï≤úÏû• Î¶¨Ïä§Ìä∏
    f.write('\n'.join(val_img_list) + '\n')

# yaml ÌååÏùº dictionary
data = {}
data['train'] = '/content/mount/MyDrive/Side_Project/Driving_Control_System/Data/train.txt'
data['val'] = '/content/mount/MyDrive/Side_Project/Driving_Control_System/Data/val.txt'
data['names'] = ['car_number']
data['nc'] = len(data['names'])

with open('/content/mount/MyDrive/Side_Project/Driving_Control_System/Data/data.yaml', 'w') as f:
    yaml.dump(data, f)
```
